```python
from qiskit import *
from qiskit.visualization import plot_bloch_multivector
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
circuit = QuantumCircuit(3)
circuit.x(0)
circuit.cx(1, 2)
# circuit.cnot(0, 1) # UPDATE: cnot() name deprecated in Qiskit 1.0
circuit.cx(0, 1)     # UPDATE: use cx() instead of cnot()
circuit.draw(output='mpl')
plt.show()
```

```python
# UPDATE: Code updated after recording to resolve breaking changes from Qiskit 1.0
circuit.remove_final_measurements()  # no measurements allowed
from qiskit.quantum_info import Statevector
statevector = Statevector(circuit)
plot_bloch_multivector(statevector)
plt.show()
```
